```python
import numpy as np
import re
from scipy import stats
import pandas as pd
import matplotlib as mpl
import seaborn as sb
```

```python
data = pd.read_csv("raw-polls.csv")
```

```python
data['type_simple'].unique()
```

```python
states = ["AL", "AK", "AZ", "AR", "CA", "CO", "CT", "DC", "DE", "FL", "GA", 
          "HI", "ID", "IL", "IN", "IA", "KS", "KY", "LA", "ME", "MD", 
          "MA", "MI", "MN", "MS", "MO", "MT", "NE", "NV", "NH", "NJ", 
          "NM", "NY", "NC", "ND", "OH", "OK", "OR", "PA", "RI", "SC", 
"SD", "TN", "TX", "UT", "VT", "VA", "WA", "WV", "WI", "WY"]
```

```python
all_state_data = {'states':states, 'districts':[], 'total_polls':[], 'percent_correct':[], 'average_error':[], 'unique_pollsters':[], 'average_bias':[]}
for i in states:
    r = re.compile(i+".*")
    li = list(filter(r.match, data['location'].unique()))
    #print(li)
    all_state_data['districts'].append(li)
    state_data = data[data['location'].isin(li)]
    #print(len(state_data))
    all_state_data['total_polls'].append(len(state_data))
    #print(len(state_data[state_data['rightcall']==1])/len(state_data)*100)
    all_state_data['percent_correct'].append(len(state_data[state_data['rightcall']==1])/len(state_data)*100)
    #print(state_data['error'].mean())
    all_state_data['average_error'].append(state_data['error'].mean())
    all_state_data['average_bias'].append(state_data['bias'].mean())
    all_state_data['unique_pollsters'].append(len(state_data['pollster'].unique()))
all_state_data = pd.DataFrame(all_state_data)
all_state_data
```

```python
race = {'type_detail': [], 'race': [], 'avg_samplesize': [], 'average_error': []}
for i in list(data['race'].unique()):
    data_simple = data[data['race']==i]
    race['average_error'].append(data_simple['error'].mean())
    race['race'].append(i)
    race['type_detail'].append(str(stats.mode(data_simple['type_detail'])[0][0]))
    race['avg_samplesize'].append(data_simple['samplesize'].mean())
race = pd.DataFrame(race)
race
```
